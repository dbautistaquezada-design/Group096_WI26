## Research Question

Has mainstream music, specifically comprising the monthly top three songs on the Billboard Hot 100 chart, become more homogenized (sonically and/or lyrically), post-streaming (2013-2026) versus pre-streaming (2000-2013)?

## Background and Prior Work

As digitalization has rapidly advanced, no stone has been left unturned. The way we listen to music has been one of the biggest changes; progressing from physical cassettes, records, and CDs to singular devices that hold hundreds of millions of tracks. This unparalleled access to music has not come lightly. Beginning in the 2010s, music consumers are able to purchase or subscribe to a streaming platform to gain access to a dataset of songs, readily available to them. Since the release of streaming platforms like Spotify and Apple Music, the barrier of entry to consume and release music has lowered exponentially. 

However, the way we consume music has also changed, with the algorithm becoming the ruling factor for many streaming platforms and social media sites. The more a song appeals to an algorithm, the more streams a song has, and all the more money it makes.  This cause and effect has led to what we believe has been a homogenization of music since the streaming age began, where songs are designed for mass appeal. In this project, we’d like to explore whether or not music has become more homogenized, sonically and/or lyrically, when comparing post-streaming (2013-2026) and pre-streaming (2000-2013) eras.

Perhaps one of the clearest signs of how music has changed is the change in the average length of music over time, which has decreased by more than 30 seconds over time.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Many of the top songs of the past few years have clocked in at under three minutes, a clear departure from longer music. This is most likely tied to the difference streaming makes when considering song structure. Researchers have found a correlation between skipping music and song structure<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) where users are likely to skip music before the song has even made it to 30 seconds. Since streaming services only pay artists for “plays” and “plays” only count if a song has been played for longer than 30 seconds,<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) it stands to reason that many artists and songwriters have focused their energy towards creating catchy music that’s interesting enough to capture attention, but short enough that listeners won’t get bored and skipping, leaving artists without their hard-earned revenue. 

In a previous study from Tilburg University, the impact of streaming services on homogenized mainstream music is mixed. In their research, Knox and Datta look at the supply and demand side of the industry, focusing on the “acoustic diversity” of a given song: the pitch, loudness, valence, tempo, and duration. In the early beginnings of platforms like Napster, an illegal music sharing platform, YouTube, and Spotify, artists were able to easily create and distribute their music to a large audience in a way that had not been seen before. Musicians tended to mimic each other in the early stages of the new widespread distribution of music, leading to an initial decrease in diversity in music quality.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) However, when streaming services like Spotify became more popularized, algorithms in place were able to provide new recommendation systems and search tools that enabled listeners to discover a vast amount of music readily available to them. From a demand standpoint, consumers have more choices of music selection available to them, thus allowing for more diversity within their consumption. 

Despite the initial dip in acoustic diversity, there has been a weak, steady increase in acoustic diversity since the release of streaming platforms due to the large catalog of recorded music available. However, our project is looking to see if special media platforms like TikTok or Instagram influence the homogeneity of music selections within streaming platforms. We aim to explore this topic by focusing on the categorization of songs through factors such as tempo, lyrics, etc, that were on the Billboard Top 100 over the last 60-70 years. Through our exploration and analysis of this data, we hope to discover whether streaming has truly homogenized popular music or if streaming has had little to no effect on the variability of popular music.

## Hypothesis


We believe that popular music has become more homogenized over time. This could be for a wide variety of reasons, but we theorize that it is largely because of two reasons. The first is because of corporate involvement with music creation. Music corporations want to maximize profits and popularity of an artist's music, which makes them push for their artists to create music that is broader, more generic, and more formulaic, which has led popular songs to sound more similar in recent years. This is compounded by the second cause, the rise of streaming services. Streaming services are operated as a business that pits artists against each other in a popularity contest. More streams means an artist's music is pushed out more, which therefore means more money for the artist and exponentially more money for the business. This also encourages streaming services to promote music that appeals to the broadest audience, as it generates the most revenue for them. 

## Data

Our ideal dataset would consist of the categorization of songs based on different factors that can be mapped as similarity vectors to one another. Such factors would consist of tempo, instrumentals, lyrics, and more. This will ideally be used in tandem with a chart of the top few songs off the Billboard Top 100 over the last 25+ years.

IDEAL DATASET: 
* Since we want to determine whether songs have become more similar, the point of comparison can be broken down to two distinct parts: audial comparison and lyrical comparison. In each we would compare specific features as follows:
    * Features of audio analysis: tempo, key, instrumentation, genre
    * Features of lyrical analysis: sentiment, explicitness, words-per-minute, chorus pattern/length

* Two separate datasets would be compiled: one for mainstream music that was released in our pre-streaming time period (2000-2013) and one for music released in our post-streaming time period (2013-2026). Each dataset would contain the monthly top three songs on the Billboard Hot 100 chart from 2000-2026 (amounting to about 936 data points) with the following metadata:
```
{
        title: "",
    	artist: "",
    	billboard_rank: "",
    	year: "",
    	month: "",
    	audio_data: 
                    {
                        tempo:"",
                        timbre: "",
                        key: "",
                        instrumentation:"", 
                        genre:""
                    		
                    },
        lyrical_data: 
                	{
                    	sentiment:"",
                    	explicitness:"", 
                    	word_per_min: "",
                    	chorus_length: "",
                    	chorus_freq: ""
                    }
}
```
This data would be collected by filtering out audio analysis datasets, lyrical breakdown datasets, and song structure datasets for the specific songs we’re comparing. 


REAL DATASET(s): 

160K Spotify Songs (1921-2020): 
https://www.kaggle.com/datasets/fcpercival/160k-spotify-songs-sorted
* This dataset includes measurements of audio features of 160,000 songs released on Spotify between 1921-2020. For the purposes of this EDA, we would extract the following data metrics for the relevant songs: for sonic comparisons: acousticness, danceability, energy, instrumentalness, loudness, tempo, key; for lyrical comparisons: explicitness. These metrics are measured on a scale from 0-1, and so their similarity to each other can be easily computed. This dataset can be directly downloaded from Kaggle for use. 

Hit Song Prediction Lyrics Dataset: 
https://github.com/Orfium/hsp-lyrics-dataset
* This dataset contains data for more then 95,000 songs describing their Billboard rank, low and high-level audio features, and text embeddings for the lyrics of each song. These text embeddings can be compared using metrics like cosine similarity to gauge the overall similarity of songs lyrics pre- and post-streaming. This dataset can be downloaded from Github. 

CCMusic Song Structure Dataset: 
https://huggingface.co/datasets/ccmusic-database/song_structure
* This dataset contains song structure annotations for around 300 pop songs, breaking down the frequency of verses and chorus segments. Determining whether a certain subset of songs from the pre- and post-streaming releases display difference in structure similarity could be useful in gauging overall similarly. This dataset can be downloaded from HuggingFace. 

Free Music Archive: A dataset for Music Analysis
https://github.com/mdeff/fma/tree/master
* This dataset consists of a comprehensive archive of songs with extracted audio features and genres. The audio features can be measured in similarity based on metrics such as spectral contrast, spectral bandwidth, and MFCC (which gauge timbre) as well as qualities like tempo and tonality. This dataset can be downloaded from Github.
